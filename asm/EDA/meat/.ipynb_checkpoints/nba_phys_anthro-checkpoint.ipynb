Hand length and Hand width: From 2010-2011\

STANDING REACH,	WEIGHT (LBS), WINGSPAN consistently show up. We probably remove weight because we already have it. For the sake of simplicity, let just merge standing reach and wingspan to the data to answer all the questions about discovery and PER. Take note that some older players may not have their andro record, so before plotting you may wanna drop them if feel like to.\
A bad news is that we don't have the draft year information in this andro. Maybe create one? It is draft combine, so it should be the same as the draft year. This is good.

So, firstly:

- create an extra column call draft year
- drop HEIGHT W/O SHOES, HEIGHT W/ SHOES, WEIGHT (LBS), POS 
- merge all andro, 
- then join outer based on draft year and name (hopefully no duplicate)

```python
import glob
files = glob.glob("nba*andro*csv")
sample = pd.read_csv(files[0])
```

```python
files.sort()
# Constructed the dropped list: must do this because of some special chars
dropped= list(sample.columns[5:7])
dropped.append(sample.columns[1])
dropped.append(sample.columns[8])

datasets = []

draft_years = range(2000, 2020)
for file_name, year in zip(files, draft_years):
    data = pd.read_csv(file_name)
    # Create new column
    data['Draft Year'] = year
    # Drop height, weight and pos
    data = data.drop(dropped, axis=1)
    datasets.append(data)
```

```python
# Union
anthro = pd.concat(datasets, ignore_index=True)
anthro.tail()
```

```python
anthro.to_csv('all_nba_anthro.csv')
```

```python
import re

r = re.compile(r"([0-9]+)' ([0-9]*\.?[0-9]+)\'\'")

def get_cm(el):
    try:
        m = r.match(el)
        if m == None:
            return float('NaN')
        else:
            return 2.54*(int(m.group(1))*12 + float(m.group(2)))
    except:
        return None
```

```python
anthro = pd.read_csv('all_nba_anthro.csv', index_col=0)
anthro.tail()
```

```python
anthro['Wingspan_cm'] = anthro['WINGSPAN'].apply(get_cm)
anthro['Standing_reach_cm'] = anthro['STANDING\xa0REACH'].apply(get_cm)
```

```python
anthro.to_csv('all_nba_anthro.csv')
```

## Joining

```python
anthro = pd.read_csv('all_nba_anthro.csv')
nba_all_physique = pd.read_csv('all_nba_phys_stats.csv')
```

```python
# Rename the PLAYER to Player, prepare for an outer join
anthro.rename(columns={'PLAYER': 'Player'}, inplace=True)
```

```python
def und(x):
    if x == 'Undrafted':
        return 0
    else:
        return x
```

```python
nba_all_physique['Draft Year'] = nba_all_physique['Draft Year'].apply(und)
```

```python
nba_all_physique['Draft Year'] = pd.to_numeric(nba_all_physique['Draft Year'])
```

```python
nba_all_physique[nba_all_physique['Player'] == 'Kobe Bryant']
```

```python
phythro = pd.merge(nba_all_physique, anthro, how='left', on=['Player', 'Draft Year'])
```

```python
phythro[phythro.duplicated('Player')]
```

```python
phythro.rename(columns={'Draft Year': 'Draft_Year', 
                        'Draft Round':'Draft_Round', 
                        'Draft Number': 'Draft_Number', 
                        'BODY\xa0FAT\xa0%': 'Body_fat',
                        '\n            HAND\xa0LENGTH\xa0(inches)\n          ': 'Hand_length',
                       'HAND\xa0WIDTH\xa0(inches)':'Hand_width',
                       'STANDING\xa0REACH':'Standing_reach',
                       'WINGSPAN':'Wingspan'},
              inplace=True)
```

```python
phythro.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
phythro.to_csv('phythro_nba.csv')
```
